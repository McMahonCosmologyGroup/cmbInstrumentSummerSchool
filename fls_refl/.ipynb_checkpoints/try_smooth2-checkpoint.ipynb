```python
from smooth import smooth_hann
import random
import numpy as np

import matplotlib.pyplot as plt
```

```python
x_pure = np.linspace(0, 100, 200)
a = [random.randrange(-10, 10) for i in range(200)]
y_noise = x_pure + a
plt.figure()
plt.plot(x_pure, y_noise)
plt.show()
```

```python
x_smooth = smooth_hann(x_pure).smooth()
y_smooth = smooth_hann(y_noise).smooth()
```

```python
plt.figure()
plt.plot(x_smooth, y_smooth)
plt.plot(x_pure, y_noise)
plt.show()
```
